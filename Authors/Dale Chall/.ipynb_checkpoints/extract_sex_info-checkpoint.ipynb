```python
import os
import json
import pandas as pd
```

```python
all_file = os.listdir(os.getcwd())
txt_file = list(filter(lambda x:x.split('.')[-1]=='txt', all_file)).sort()
```

```python
all_authors = []
for tf in txt_file:
    with open(tf, 'r') as f:
        alphabet = tf[7]
        contents = f.read()
        contents = contents.replace('<script type="text/javascript">generateChartData(','')
        contents = contents.replace(');</script>','')
        authors_dat = eval(contents)
        all_authors += authors_dat
```

```python
sex_data = []
for d in all_authors:
    name = d['AuthorName']
    sex = d['Sex']
    sex_data.append([name, sex])
```

```python
sex_df = pd.DataFrame(sex_data, columns=["Author Name", "Sex"])
```

```python
sex_df.to_csv("author_sex.csv", index = False)
```
