```python
import pandas as pd
```

```python
RES_df = pd.read_excel('RES.xlsx',engine='openpyxl', index=None)
RES_df = RES_df.iloc[:, 1:]
```

```python
RES_df[['Journal','Volume','Issue','Date','Page','Source URL']] = RES_df['Issue'].str.split(', ', expand = True)
RES_df[['Month', 'Year']] = RES_df['Date'].str.split(' ', expand = True)
```

```python
RES_df
```

```python
RES_df.drop('Journal', axis=1, inplace=True)
```

```python
## handle author
author_col = [f'Authors{i}' for i in range(1, 8)]
for col in author_col:
    RES_df[col] = RES_df[col].astype(str).str.replace(r'\s+and\s+|,(?!\s*Jr\.)\s*', '', regex=True)
```

```python
RES_df = RES_df.replace('nan','')
```

```python
## re-order
order =[
    "Source URL",
    "Title",
    "Volume",
    "Issue",
    "Month",
    "Year",
    "Abstract"
] + author_col
```

```python
RES_df = RES_df[order]
```

```python
RES_df
```

```python
RES_df.to_csv('RES_parsed.csv', index=None)
```
