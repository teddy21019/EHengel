```python
import pandas as pd
```

```python
QJE_df = pd.read_excel('QJE_need_parse.xlsx',engine='openpyxl', index=None)
QJE_df['Issue']
```

```python
QJE_df[['Volume','Issue','Date','Page','Source URL']] = QJE_df['Issue'].str.split(', ', expand = True)
QJE_df[['Month', 'Year']] = QJE_df['Date'].str.split(' ', expand = True)
```

```python
QJE_df['Authors1']
```

```python
## handle author
author_col = [f'Authors{i}' for i in range(1, 8)]
for col in author_col:
    QJE_df[col] = QJE_df[col].astype(str).str.replace(r'\s+and\s+|,(?!\s*Jr\.)\s*', '', regex=True)
```

```python
QJE_df = QJE_df.replace('nan','')
```

```python
## re-order
order =[
    "Source URL",
    "Title",
    "Volume",
    "Issue",
    "Month",
    "Year",
    "Abstract"
] + author_col
```

```python
QJE_df = QJE_df[order]
```

```python
QJE_df
```

```python
QJE_df.to_csv('QJE_parsed.csv', index=None)
```
